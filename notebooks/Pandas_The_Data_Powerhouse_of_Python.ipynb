**What Is a Series?**
<br> A Series is like a single column of data.

```python
import pandas as pd
sales = pd.Series([1000, 1200, 1150, 1400])
print(sales)
```

**What Is a DataFrame?**
<br> A DataFrame is like a full table with rows and columns.

```python
data = {
    "Month": ["Jan", "Feb", "Mar"],
    "Sales": [1000, 1200, 1100]
}
df = pd.DataFrame(data)
print(df)
```

**Accessing DataFrames**
<br> You can easily view specific parts of your DataFrame:

```python
print(df["Sales"])      # Access a column
print(df.head(2))       # View first 2 rows
print(df.tail(1))       # View last row
```

**Accessing and Modifying DataFrames**
<br> Using loc (label-based)

```python
print(df.loc[0])  # Row with index 0
print(df.loc[0, "Sales"])  # Sales value in row 0
```

Using iloc (position-based)

```python
print(df.iloc[1])  # 2nd row (position 1)
print(df.iloc[1, 1])  # 2nd row, 2nd column
```

**Pandas Functions: Powerful Shortcuts**

```python
print(df.describe())   # Summary stats
print(df["Sales"].mean())  # Average sales
print(df["Sales"].max())   # Max value
print(df["Sales"].sum())   # Total sales
```
