**What Is a NumPy Array?**

```python
import numpy as np
sales = np.array([1200, 1300, 1250, 1400, 1350])
print(sales)
```

**Calculating Business Metrics**

Mean (Average)

```python
average = np.mean(sales)
print("Average monthly sales:", average)
```

Total Sales

```python
total = np.sum(sales)
print("Total sales so far:", total)
```

**Slicing Arrays**

```python
print(sales[1:4])
```

**Business Scenario: Branch Revenue Table**

```python
revenue = np.array([
    [5000, 5200, 5100],  # Branch A
    [4800, 4900, 4950],  # Branch B
    [5300, 5400, 5500]   # Branch C
])
```

What is the average revenue per branch?

```python
branch_avg = np.mean(revenue, axis=1)
print(branch_avg)
```

What is the highest revenue in any month?

```python
max_revenue = np.max(revenue)
print(max_revenue)
```
