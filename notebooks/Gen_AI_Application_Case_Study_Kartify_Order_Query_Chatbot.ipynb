## **Strategy Selection**

image.png

Think of this like customizing a digital assistant for your business. The core AI model is like hiring a smart, well-educated employee they know a lot in general, but they dont yet understand your companys tone, preferences, or specific goals. If you ask them a basic question, theyll answer, but not always in the way you want.
**Thats where strategy selection comes in.**

This process can involve different strategies, depending on the problem we're trying to solve:

* **Prompt Engineering:** The simplest and fastest method. Think of this like giving very clear instructions before asking for help. Here, we tweak the prompt to include structure, examples, tone, or constraints.

* **Retrieval-Augmented Generation (RAG):** Useful when responses need to be grounded in external or dynamic information. RAG retrieves relevant content from documents, databases, or APIs and feeds it into the prompt, helping the AI generate factually accurate and up-to-date responses.

* **Agentic AI:** In more complex cases, we need the AI to perform reasoning, tool use, and coordinated steps like looking up order details, applying return policies, or calculating delivery timelines. Agentic AI enables multiple specialized agents to work together to generate accurate, context-aware responses.

### Answer Generation Agent

This tool in the Agentic AI workflow crafts personalized responses for the user based on their query and database.

```python
def answer_generation_tool(input: str) -> str:
    prompt = f"""You are a polite and formal customer assistant replying to a user query to customer about return, replacement, delivery, cancellation, or address change.

Use the context provided below:
{input}

Guidelines:
- Respond in a short, formal, and factual tone.
- Do **not** add extra details that were not asked in the user's query.
- Do **not** format like an email  avoid greetings, sign-offs, or explanations.
- Do not offer return/replacement windows unless directly asked.
- Do not mention cancellation or refund policies unless cancellation is requested.
- Do not mention address change policy until asked.
- If the product has **not been delivered**, mention that return/replacement cannot be processed yet.
- If the product **has been delivered**, calculate based on delivery date whether return or replacement is still allowed.
- If the order has already **shipped**, reply that address changes are not possible.
- If any required info is missing, say so politely that i am connecting to human support.
- If you do not what to answer now, say so politely that i am connecting to human support.
- Never ask for any bank details

Output:
- Return only a single, relevant customer-facing message  no system instructions, reasoning, or metadata.
"""

    return llm([HumanMessage(content=prompt)]).content.strip()
```

```python
tools = [Tool(name="AnswerGenerator", func=answer_generation_tool, description="Craft response based on input.")]
```

```python
#Initialize order Agent
order_agent = initialize_agent(
    tools,
    llm,
    agent=AgentType.OPENAI_FUNCTIONS,
    verbose=False,
    handle_parsing_errors=True
)
```

```python
# Generating a response from the Agent model
query=f""" Context:{order_details}

Customer Query: Where is my order with order ID O40327?

Use tools to reply."""

order_agent.run(query)
```

## **Multi-Agent AI Workflow Implementation**

### Policy Agent

This tool within the Agentic AI workflow is responsible for calculating the eligibility for return and replacement.

```python
def policy_tool_func(input: str) -> str:
    prompt = f"""Only respond about return or replacement if the user has explicitly asked about it in their query.

Use the following context from order, shipment, and product policy data:
{input}

Your task (only if return or replacement is mentioned):
1. Check eligibility based on `actual_delivery` and product policy:
   - If `return_days_allowed` is 0, clearly state the product is not eligible for return.
   - If within window, mention last date allowed for return and replacement.
   - If the window has passed, say so.
2. Mention return conditions (e.g., Sealed only).
3. If `actual_delivery` is null, return/replacement is not yet applicable.
4. If any required info is missing, say so politely that i am connecting to human support.

"""

    return llm([HumanMessage(content=prompt)]).content.strip()
```

```python

tools = [
    Tool(name="PolicyChecker", func=policy_tool_func, description="Check return and replacement eligibility."),
    Tool(name="AnswerGenerator", func=answer_generation_tool, description="Craft final response.")
]
```

```python
#Initialize order Agent

order_agent = initialize_agent(
    tools,
    llm,
    agent=AgentType.OPENAI_FUNCTIONS,
    verbose=False,
    handle_parsing_errors=True
)
```

```python
# Generating a response from Multi Agent Model
query=f""" Context:{order_details}

Customer Query: Where is my order with order ID O40327?

Use tools to reply."""
```

```python
order_agent.run(query)
```

## **Output Evaluation**

Even if weve built a good system, its important to check how well its working. Evaluation helps us:

* See if the output is relevant and meaningful

* Ensure it is based on the right information

* Identify any errors or vague responses

* Improve the overall system for future queries



In our scenario, well now evaluate the model output with a focus on:

- **Groundedness**  Does the response stay strictly within the provided context?
- **Precision**  Does the response directly and accurately answer the users query?


image.png

### **Multi Agent AI Workflow with Output Evaluation**

This function evaluates whether the response precisely addresses the users query and grounded in the retrieved context.

```python
def evaluate_response_quality(context: str, query: str, response: str) -> dict:
    prompt = f"""
     Evaluate the assistant's response to a customer query using the provided order context.

    Context: {context}
    Customer Query: {query}
    Assistant's Response: {response}

     Instructions:
      1. **Groundedness (0.0 to 1.0)**: Score based on how well the response is factually supported by the context.
                                      - Score closer to 1 if all facts are accurate and derived from the context.
                                      - Score closer to 0 if there is hallucination, guesswork, or any fabricated information.

      2. **Precision (0.0 to 1.0)**: Score based on how directly and accurately the assistant addresses the query.
                                      - Score closer to 1 if the response is concise, focused, and answers the exact user query.
                                      - Score closer to 0 if it includes irrelevant details or misses the main point.

        Output format (JSON only):

           groundedness: float between 0 and 1 ,
             precision: float between 0 and 1

        Only return the JSON. No explanations.

      """
    score = llm.predict(prompt).strip()
    try:
      return eval(score)
    except:
      return {"groundedness": 0.0, "precision": 0.0}
```

## **Guardrailing**

image.png

As Generative AI becomes more powerful, it also becomes more important that it's used sensitively. A small change in the prompt or a vague question can sometimes lead to misleading, unsafe, or off-topic responses. Thats why we need something called **guardrails** - clear rules and checks that guide the model's behavior.

Without guardrails, even  a well-trained and optimized model might:

- Make up facts (hallucinate)
- Give incomplete or unsafe advice
- Respond inappropriately or go off-topic
- Get manipulated through tricky or misleading inputs

In short, **guardrails help make AI safe, accurate, and usable in real-world business scenarios.**

In this solution, we are applying **two layers of guardrails** to ensure that the AI assistant behaves safely, respectfully, and stays focused on the users intent throughout the interaction.

1. **Response-Level Guardrail**  Acts in real time to block unsafe, inappropriate, or misleading replies before they reach the user.

2. **Conversation-Level Guardrail**  Monitors the full chat to detect repetitive, irrelevant, or misaligned responses, flagging issues that could harm the user experience.


## Guardrailing

### **Response-Level Guardrail: `output_guard_check`**


This guardrail acts like a **real-time safety scanner** for each individual response generated by the assistant. As soon as the assistant replies, the response is checked for:

* Sensitive or risky requests (like asking for bank details or OTPs)
* Unsafe or inappropriate language
* Redirection phrases like Please contact customer service that might frustrate the user
* Escalation language or signs of miscommunication

If any of these red flags are found, the response is **blocked immediately** before it reaches the user.

```python
def output_guard_check(model_output: str) -> str:
    prompt = f"""
You are a content safety assistant. Your task is to classify if the assistant's response is appropriate.
If the message contains:
- Requests for bank details, OTPs, account numbers
- Harassment or offensive tone
- Privacy concerns or unsafe advice
- Misunderstanding and miscommunication word
- Phrases like "please contact customer service" or redirection to a human agent
- Escalated this to our support team
Return: BLOCK
Otherwise, return: SAFE
Response: {model_output}
Output:
"""
    return llm.predict(prompt).strip()
```

## Guardrailing (continued)

### **Conversation-Level Guardrail: `conversation_guard_check`**

This second guardrail takes a **holistic view of the entire conversation**. It monitors the assistants behavior over time to detect:

* Repetitive or irrelevant information (e.g., constantly offering return options when not asked)
* Situations where the assistant overshares or misunderstands the user's real intent
* Missed emotional cues or ignoring a user's explicit refusal or correction

If any of these patterns are observed, the system flags the conversation and returns a **BLOCK** to prevent poor customer experience.

```python
def conversation_guard_check(history) -> str:
    chat_summary = "\n".join([f"Customer: {h['user']}\nAssistant: {h['assistant']}" for h in history])
    prompt = f"""
You are a conversation monitor AI. Review the entire conversation and classify if the assistant:
- Repeatedly offered unnecessary return or replacement steps
- Gave more than what the user asked
- Missed signs of customer distress
- Ignored user's refusal of an option
If any of the above are TRUE, return BLOCK
Else, return SAFE
Conversation:
{chat_summary}
Output:
"""
    return llm.predict(prompt).strip()
```

```python
def chatagent():
  human = 0
  history = []
  order_id = input("Enter Order ID: ")

  order_context_raw = sqlite_agent.invoke(
      f"""
      Fetch all columns for order ID {order_id}
      """
  )

  order_info = f"Order ID: {order_id}\n{order_context_raw}\nDate: 25 July"
  response = ""

  print("\nHow can I help you\n")

  while True:
      user_query = input("Customer: ")
      if user_query.lower() in ["exit", "quit"]:
          print("Assistant: Thank you! Have a great day.")
          break

      prompt=f"""You are an intent classifier for customer service queries. Your task is to classify the user's query into one of the following 3 categories based on tone, completeness, and content.

              Return **only the numeric category ID (0, 1, 2)** as the output. Do not include any explanation or extra text.

              ### Categories:

              0  **Escalation**
                - The user is very angry, frustrated, or upset.
                - Uses strong emotional language (e.g., This is unacceptable, Worst service ever, Im tired of this, I want a human now).
                - Requires **immediate human handoff**.
                - Escalation confidence must be very high (90% or more).

              1  **Exit**
                - The user is ending the conversation or expressing satisfaction.
                - Phrases like Thanks, Got it, Okay, Resolved, Never mind.
                - No further action is required.

              2  **Process**
                - The query is clear and well-formed.
                - Contains enough detail to act on (e.g., mentions order ID, issue, date).
                - Language is polite or neutral; the query is actionable.
                - Proceed with normal handling.

              3 - **Random Question**
                - If user asked something not related to order
                - Example - What is NLP
                  ---

                Your job:
                Read the user query and return just the category number (0, 1, 2, or 3). Do not include explanations, formatting, or any text beyond the number.

                  User Query: """+user_query
      res = llm.predict(prompt).strip()

      if res == "0":
          print("Assistant: Sorry for the inconvenience caused to you. Your request is being routed to a customer support specialist for further assistance. A human agent will connect with you shortly.")
          human = 1
          break
      elif res == "1":
          print("Assistant: Thank you! I hope I was able to help with your query.")
          break

      elif res == "3":
          print("Assistant: Apologies, Im currently only able to help with information about your placed orders. Please let me know how I can assist you with those!")
          human = 1
          break

      full_prompt = f"""
        Context: {order_info}

        Customer Query: {user_query}

        Previous response: {history}

        Use tools to reply.
        """

      # Step 1: Initial agent response
      raw_response = order_agent.run(full_prompt)

      # Step 2: Evaluate quality
      initial_scores = evaluate_response_quality(order_info, user_query, raw_response)

      if initial_scores["groundedness"] < 0.75 or initial_scores["precision"] < 0.75:
          regenerated_response = order_agent.run(full_prompt)
          retry_scores = evaluate_response_quality(order_info, user_query, regenerated_response)

          if retry_scores["groundedness"] >= 0.75 and retry_scores["precision"] >= 0.75:
              final_response = regenerated_response
          else:
              final_response = "Your request is being forwarded to a customer support specialist. A human agent will assist you shortly."
      else:
          final_response = raw_response

      # Step 3: Output-level guard check
      if "forwarded to a customer support specialist" not in final_response:
          if output_guard_check(final_response) == "BLOCK":
              final_response = "Your request is being forwarded to a customer support specialist. A human agent will assist you shortly."

      # Save to chat history
      history.append({"user": user_query, "assistant": final_response})

      # Step 4: Conversation-level guard check
      if conversation_guard_check(history) == "BLOCK":
          print("Assistant: Your request is being forwarded to a customer support specialist. A human agent will assist you shortly.")
          break

      print("Assistant:", final_response)
```

#### 2. Create a New Space

   * Navigate to [Hugging face Spaces ](https://huggingface.co/spaces).
   * Click **Create New Space**.
   * Fill in:

     * Name for your Space.
     * Space SDK: Select **Docker**.
     * Choose a Docker template: Select **Streamlit**
     * Visibility: Choose *Public* or *Private*.
   * Click **Create Space**.

#### 3. Upload Your Files




   * In the new Space, click the **Files** tab.
   * Delete existing requirements.txt , DockerFile
   * Click Contribute then Upload files and add:

     * `app.py`
     * `requirements.txt`
     * `config.json`
     * `DockerFile`
     * `kartify.db`
   * Commit the upload.

**NOTE :** Paste your API key into the `config.json` file, save it, and then upload the updated file.


#### 4. Build and Launch



   * Hugging Face will automatically detect the `Dockerfile` and start building the container.
   * Wait a few minutes for the build to complete and the app to go live.

#### 5. Access and Test

* Go to the App tab or the Space URL to view and test your running Streamlit app.
