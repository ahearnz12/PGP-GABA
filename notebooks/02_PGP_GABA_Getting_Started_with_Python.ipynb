### 1.4 Querying Inventory Insights

Now that our inventory is well structured, we would want to extract basic insights using Python.

*   We now want to know the number of items available for each product in one glance.



**Prompt:** <font size=3 color="blue"> Based on the above created lists, show how many units are available for each product.

````python
# prompt: Based on the above created lists, show how many units are available for each product.

# We can use a dictionary to store the number of units available for each product.
items = ['eggs', 'apples', 'mobile', 'sofa']
quantity = [5, 10, 15, 20]

# Create a dictionary to store the item and its quantity
product_availability = {}
for i in range(len(items)):
    product_availability[items[i]] = quantity[i]

# Show the number of units available for each product
product_availability
````

**As we can see**, the code written above is using **For** loop and final output is in a Python ***Dictionary*** format.
<br> In Python, using a **dictionary** is perfect when:

### 1.5 Filtering Data Based on Conditions

To maintain a smooth inventory process, it's important to ensure that all items are adequately stocked.

````python
# Create a dictionary with item names as keys and their quantities as values
product_availability = {
    'eggs': 5,
    'apples': 10,
    'mobile': 15,
    'sofa': 2
}
````

WOW Superstore regularly runs promotional campaigns across various categories to attract more customers. One of the goals of the store managers is to track heavily discounted products

* Let’s say the discount data for a few products has been recorded in a dictionary provided below. We want to **identify the heavily discounted products**. This can be done by using "**Conditional Statements**"




**What are Conditional Statements?**

> Conditional statements are an important part of decision-making in Python. Results are generated for the satisfied condition.

**If Else Conditional Statement**

> The if-else statement in Python is used for decision-making. The general syntax of an if-else statement is:
```
 if (test expression):

 ```
>It contains a body of code which runs only when the condition given in the if statement is true. If the condition is false, then the optional else statement runs which contains some code for the else condition.


Now, let's define the dictionary and write the prompt to get the desired outcome.

````python
discounts = {
    'binders': 0.2,
    'appliances': 0.8,
    'bookcases': 0.5,
    'chairs': 0.3
}
````

**Prompt:** <font size=3 color ="blue"> Identify the products from the above that are currently being sold at a discount of 50% or more.

````python
# prompt: Identify the products from the above that are currently being sold at a discount of 50% or more.

discounts = {
    'binders': 0.2,
    'appliances': 0.8,
    'bookcases': 0.5,
    'chairs': 0.3
}

# Identify products with a discount of 50% or more
heavily_discounted_products = []
for product, discount in discounts.items():
  if discount >= 0.5:
    heavily_discounted_products.append(product)

print(f"The following products have a discount of 50% or more: {heavily_discounted_products}")
````

We just wrote a code to identify heavily discounted products. This works perfectly- but notice how we had to:

* Create an empty list,
* Loop through the dictionary,
* Check the condition, and
* Append the result.

Python offers a more concise way to write the same logic in a single line using **list comprehension**. It does the exact same thing as the for loop above — but is shorter and often easier to read.





**Prompt:** <font size = 3 color = "blue"> Identify the products from the above that are currently being sold at a discount of 50% or more. Use list comprehension technique.

````python
# prompt: Identify the products from the above that are currently being sold at a discount of 50% or more. Use list comprehension technique.

discounts = {
    'binders': 0.2,
    'appliances': 0.8,
    'bookcases': 0.5,
    'chairs': 0.3
}

heavily_discounted_products = [product for product, discount in discounts.items() if discount >= 0.5]

print(f"The following products have a discount of 50% or more: {heavily_discounted_products}")
````

Based on the above, create a Python function called get_category that takes an item_name as input. If the item exists, it returns the item and corresponding category in a list. If the item is not found, it returns the item and "Item not found" text in a list.

````python
# prompt: Based on the above, create a Python function called get_category that takes an item_name as input. If the item exists, it returns the item and corresponding category in a list. If the item is not found, it returns the item and "Item not found" text in a list.

def get_category(item_name):
  items = ['eggs', 'apples', 'mobile', 'sofa']
  categories = ['Food', 'Food', 'Electronics', 'Furniture']

  try:
    index = items.index(item_name)
    return [item_name, categories[index]]
  except ValueError:
    return [item_name, "Item not found"]
````

To understand the above code better, let write a prompt to generate code explanations for each step done in the code.

**Prompt:** <font size = 3 color = "blue"> Re-write  the above code with explanations for each steps.

````python
# prompt: Re-write  the above code with explanations for each steps.

# Function to get the category of an item
def get_category(item_name):
  """
  This function takes an item name as input and returns its category.

  Args:
    item_name: The name of the item to search for.

  Returns:
    A list containing the item name and its category.
    If the item is not found, it returns "Item not found".
  """
  items = ['eggs', 'apples', 'mobile', 'sofa']  # List of items
  categories = ['Food', 'Food', 'Electronics', 'Furniture']  # Corresponding categories

  try:
    # Try to find the index of the item in the items list
    index = items.index(item_name)
    # If the item is found, return the item and its category
    return [item_name, categories[index]]
  except ValueError:
    # If the item is not found, return "Item not found"
    return [item_name, "Item not found"]

# Example usage
print(get_category("mobile"))  # Output: ['mobile', 'Electronics']
print(get_category("chair"))  # Output: ['chair', 'Item not found']
````
